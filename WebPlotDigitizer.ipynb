```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
defa = pd.read_csv(("Default.csv"), sep=';',header=None)
xx = np.array( defa[0].str.replace(',','.').astype(float))
yy = np.array( defa[1].str.replace(',','.').astype(float))
```

```python
plt.figure(figsize=(12,7))
plt.plot(xx,yy,'o')
plt.show()
```
